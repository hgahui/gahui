```python
df = '/content/mydrive/MyDrive/DataSet/datajoin/joindata_final.csv'
jdf = pd.read_csv(df)
jdf
```

```python
jdf.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
jdf
```

```python
jdf.rename(columns={'만인당기관수(개소)':'만인당은행수(개소)'}, inplace=True)
jdf.rename(columns={'합계(개소)':'은행수합계(개소)'}, inplace=True)
jdf.rename(columns={'합계':'식품업합계(개소)'}, inplace=True)
jdf.rename(columns={'합계.1':'호텔숙박업합계(개소)'}, inplace=True)
jdf.rename(columns={'합계_소계':'지방세_합계'}, inplace=True)
```

```python
columns = list(jdf.columns)
```

```python
def getCatCol(data):
    cat_columns = []
    for col in data.columns:
        if data[col].dtype == 'object':
            cat_columns.append(col)
    return cat_columns
```

```python
cat_columns = getCatCol(jdf)
```

```python
num_columns = [x for x in columns if x not in cat_columns]
```

```python
print(columns)
print(cat_columns)
print(num_columns)
```

```python
num_columns_new = [i for i in num_columns if i not in removers]
```

```python
num_columns_new
```

```python
plt.figure(figsize = (20, 20))
for i in range(len(num_columns_new)):
  plt.subplot(4, 4, i+1)
  sns.histplot(data = jdf, x = num_columns_new[i],
      bins = 10,
      kde = True,
      );
```

```python
plt.figure(figsize = (40, 40))
for i in range(len(num_columns_new)):
  ax = plt.subplot(4, 4, i+1)
  ax = sns.boxplot(data = jdf, x = "행정구역별(1)", y = num_columns_new[i])
```

```python
plt.figure(figsize = (40, 40))
for i in range(len(num_columns_new)):
  ax = plt.subplot(4, 4, i+1)
  ax = sns.scatterplot(data = jdf, x = "행정구역별(1)", y = num_columns_new[i])
```

```python
newjdf = jdf[num_columns_new]
```

```python
ax = sns.PairGrid(jdf, diag_sharey=False, aspect=1.2, hue = '행정구역별(1)')
ax.map_upper(sns.scatterplot)
ax.map_lower(sns.kdeplot)
ax.map_diag(sns.histplot)
```

```python
ax = sns.PairGrid(newjdf, diag_sharey=False, aspect=1.2)
ax.map_upper(sns.scatterplot)
ax.map_lower(sns.kdeplot)
ax.map_diag(sns.histplot)
```

```python
corr_mat = np.round(jdf[num_columns_new].corr(), 4)
corr_mat_long = corr_mat.stack().reset_index(name="correlation")
```

```python
plt.figure(figsize=(8, 8))

ax = sns.scatterplot(data=corr_mat_long, x="level_0", y="level_1", hue="correlation", palette="vlag", hue_norm=(-1, 1), edgecolor=".7", sizes=(50, 250), size="correlation")
ax.set(xlabel="", ylabel="", aspect="equal")
for label in ax.get_xticklabels():
  label.set_rotation(90)
```

```python
plt.figure(figsize=(10, 10))
heatmap = sns.heatmap(corr_mat, vmin = -1, vmax = 1, annot = True, cmap = "coolwarm")
plt.show()
```
