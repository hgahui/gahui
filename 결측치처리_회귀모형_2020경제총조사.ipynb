```python
# 값이 null인 행의 인덱스 리스트
null_indices = data[data['임차료'].isnull()].index.tolist()

# 값이 null이 아닌 행의 인덱스 리스트
nonnull_indices = data[data['임차료'].notnull()].index.tolist()
```

```python
data['총합계_연간급여'] = data['총합계_연간급여'].fillna(0)
```

```python
data2 = data.drop(null_indices, axis=0)
```

```python
X1 = np.array(data2['총합계_연간급여']).reshape(-1, 1)
X2 = np.array(data2['매출액']).reshape(-1, 1)
X3 = np.array(data2['주사업_수익액']).reshape(-1, 1)
Y = np.array(nonnull_indices).reshape(-1, 1)

X = np.concatenate((X1, X2, X3), axis=1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X, Y)
```

```python
print(lr.coef_[0])
print(lr.intercept_)
```

```python
import statsmodels.api as sm
results = sm.OLS(Y, sm.add_constant(X)).fit()
results.summary()
```

```python
def lr_value(i):
    y = 399257.112 - 99.975 * data.at[i, '총합계_연간급여'] + 3.349 * data.at[i, '매출액'] - 0.231 * data.at[i, '주사업_수익액']
    return y
```

```python
for i in tqdm(range(len(data['임차료']))):
    if pd.notna(data.at[i, '임차료']) != True:
        data.at[i, '임차료'] = lr_value(i)
```

```python
data['임차료'].describe()
```

```python
# 값이 null인 행의 인덱스 리스트
null_indices_ad = data[data['광고선전비'].isnull()].index.tolist()

# 값이 null이 아닌 행의 인덱스 리스트
nonnull_indices_ad = data[data['광고선전비'].notnull()].index.tolist()
```

```python
data3 = data.drop(null_indices_ad, axis=0)
```

```python
X11 = np.array(data3['총합계_연간급여']).reshape(-1, 1)
X22 = np.array(data3['매출액']).reshape(-1, 1)
X33 = np.array(data3['주사업_수익액']).reshape(-1, 1)
YY = np.array(nonnull_indices_ad).reshape(-1, 1)

XX = np.concatenate((X11, X22, X33), axis=1)
```

```python
lr.fit(XX, YY)
```

```python
print(lr.coef_[0])
print(lr.intercept_)
```

```python
import statsmodels.api as sm
results = sm.OLS(YY, sm.add_constant(XX)).fit()
results.summary()
```

```python
def lr_value_ad(i):
    y = 380701.697 - 89.024 * data.at[i, '총합계_연간급여'] - 0.695 * data.at[i, '매출액'] + 6.348 * data.at[i, '주사업_수익액']
    return y
```

```python
for i in tqdm(range(len(data['광고선전비']))):
    if pd.notna(data.at[i, '광고선전비']) != True:
        data.at[i, '광고선전비'] = lr_value_ad(i)
```
