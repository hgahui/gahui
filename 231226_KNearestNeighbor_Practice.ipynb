```python
from sklearn.datasets import load_breast_cancer

breast_cancer_data = load_breast_cancer()
```

```python
import pandas as pd

df_data = pd.DataFrame(breast_cancer_data.data)                 # 데이터를 돌려준다.
df_labels = pd.DataFrame(breast_cancer_data.target)             # 데이터의 라벨을 돌려준다.
```

```python
df_labels.head()
```

```python
print(breast_cancer_data.target_names)       # malignant - 음성, benign - 양성
```

```python
df_data.head()
```

```python
df_data.describe()             # 항목별 최대/최소 및 기초 통계량 확인
```

```python
# K-Nearest Neighbor 알고리즘 적용 전 최소-최대 정규화 함수 정의
def min_max_normalization(lst):
  normalized = []

  for value in lst:
    normalized_num = (value-min(lst)) / (max(lst)-min(lst))
    normalized.append(normalized_num)

  return normalized

# 최소-최대 정규화 함수로 정규화
for x in range(len(df_data.columns)):
  df_data[x] = min_max_normalization(df_data[x])

df_data.describe()
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_labels, test_labels = train_test_split(df_data, df_labels, test_size = 0.2, random_state = 100)
```

```python
print(len(train_data))
print(len(train_labels))
print(len(test_data))
print(len(test_labels))
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors =3)
```

```python
classifier.fit(train_data, train_labels)
```

```python
classifier.score(test_data, test_labels)         # k=3일 때의 모델의 정확도(Accuracy) 평가
```

```python
# k=1부터 k=100까지 모델의 정확도 저장 후 시각화

import matplotlib.pyplot as plt

k_list = range(1, 101)
accuracies = []

for k in k_list:
  classifier = KNeighborsClassifier(n_neighbors = k)
  classifier.fit(train_data, train_labels.values.ravel())
  accuracies.append(classifier.score(test_data, test_labels))

plt.plot(k_list, accuracies)
plt.show()
```
